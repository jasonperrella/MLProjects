Different Types of Plots in Matplotlib

Line Chart

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,3,2,5])
plt.plot(x,y, linestyle = 'dotted')
plt.show()
```

Bar Chart

```python
x2 = np.array(["X", "Y", "Z", "W"])
y2 = np.array([25,30,27,22])
plt.bar(x2,y2)
plt.show()
```

Scatter Plot

```python
x3 = np.array([20,15,11,7,8,10])
y3 = np.array([2,7,15,8,10,12])
plt.scatter(x3,y3)
```

Pie Chart

```python
p = np.array([50,50,50,50])
plt.pie(p)
```

Histogram

```python
h1 = np.random.normal(170,10,250)
plt.hist(h1)
```

# SEABORN

```python
import seaborn as sns
# AND MATPLOTLIB as plt
# And pandas as pd

# Boxplot
# Distplot
# Regplot

#Use seaborn built in data set
df = sns.load_dataset("tips")
df
```

Boxplot

```python
df.boxplot(by = 'day', column = ['total_bill'], grid = True)
```

Distplot

```python
titanic = sns.load_dataset("titanic")
titanic.head()
age1 = titanic['age'].dropna()
sns.displot(age1, bins=50, kde=False)
plt.show()
```

Regplot

```python
data = sns.load_dataset("mpg")
```

```python
data.head()
```

```python
sns.regplot(x='mpg', y='acceleration', data=data)
plt.show()
```
